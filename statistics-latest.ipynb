```python
def read_file(f):
    data = {}
    with open(f, "r+") as file:
        all_lines = file.readlines()
    for i in range(0,len(all_lines)):
        if "CONTAINER ID" not in all_lines[i]:
            line_values = all_lines[i].split(" ")
            line_values = [x for x in line_values if x != '']
            if len(line_values)>=10:
                name = line_values[1]
                if 'spark' in name:
                    if name not in data.keys():
                        data[name] = {}
                    if 'network_i' not in data[name].keys():
                        data[name]['network_i'] = []
                    data[name]['network_i'].append(line_values[7])
                    if 'network_o' not in data[name].keys():
                        data[name]['network_o'] = []
                    data[name]['network_o'].append(line_values[9])
    return data

def get_network_activity(data):
    s_i = 0.0
    for each_container in data.keys():
        for x in data[each_container]['network_i']:
            if len(x)>2 and x[-1] == 'B' and x[-2] == 'M':
                s_i += float(x[:-2])
    s_o = 0.0
    for each_container in data.keys():
        for x in data[each_container]['network_o']:
            if len(x)>2 and x[-1] == 'B' and x[-2] == 'M':
                s_o += float(x[:-2])
    return s_i, s_o
```

```python
passes = ["1pass", "2pass", "3pass", "4pass", "5pass", "all"]
```

```python
for num_workers in [2, 3, 4, 5, 6]:
    for v in [5,4,3,2,1,0]:
        print("num workers - ", num_workers, " passes - ", passes[v], end=" ")
        try:
            docker_stats_file = read_file("results/"+ str(num_workers) +"workers/" + passes[v] + "-reduce.csv")
            netin, netout = get_network_activity(docker_stats_file)
            total_net = netin + netout
            total_net = total_net/1024/1024
            print(total_net)
        except Exception as e:
            print("---")
    print()
```

```python
all_reduce_5 = read_file("comp/all-reduce.csv")
pass1_itrs12 = read_file("comp/1pass-reduce.csv")
pass1_itrs18 = read_file("comp/1pass-18-reduce.csv")

y1 = []
y2 = []

net_allreduce = get_network_activity(all_reduce_5)
y1.append(net_allreduce[0])
y2.append(net_allreduce[1])

net_1 = get_network_activity(pass1_itrs12)
y1.append(net_1[0])
y2.append(net_1[1])

net_2 = get_network_activity(pass1_itrs18)
y1.append(net_2[0])
y2.append(net_2[1])
```

```python
import matplotlib.pyplot as plt
labels = ['all_reduce(12 itrs)', '1 pass(12 itrs)', '1 pass(18 itrs)']

width = 0.35       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()
# ax.set_yscale('log')
ax.bar(labels, y1, width, label='Network(I)')
b = ax.bar(labels, y2, width, bottom=y1, label='Network(O)')

accuracies = ["91.59%", "33.46%", "89.90%"]
for i in range(len(accuracies)):
    height = b[i].get_height()
    ax.text(x = b[i].get_x() + b[i].get_width() / 2, y = height * 1.8, s = accuracies[i], ha='center')

ax.set_ylabel('Overall Network Activity (MB)')
ax.set_title('Horovod Network Overhead')
ax.legend()

plt.xticks(rotation=45, ha='right')
plt.show()
```
